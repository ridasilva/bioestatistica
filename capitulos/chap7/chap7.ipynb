## Introdução
No último capítulo, você aprendeu a usar a variável padrão z como estatística da média amostral para testar uma hipótese nula. Para calcular o z para uma média amostral, você deve conhecer o desvio padrão da população ($\sigma$). Porém, na maioria das situações de pesquisa, o desvio padrão da população não é conhecido. Neste capítulo, você aprenderá como usar a estatística t de amostra única quando não souber o desvio padrão da população.

Assim como o z para uma média amostral, a estatística t de amostra única é usada para determinar se a diferença entre uma média amostral e uma média populacional é provavelmente devida a um erro amostral. O z para uma média amostral e o t de amostra única são diferentes porque o teste t de amostra única usa o desvio padrão da amostra (DP) em vez do desvio padrão da população para calcular uma estimativa do erro amostral esperado. Se você conhece o desvio padrão populacional ($\sigma$), você deve sempre fazer um z para uma média amostral, mas quando não o conhece, é necessário fazer um teste t de amostra única. Quase todo o resto é o mesmo entre o t de amostra única e o z para uma média amostral.

## Informações conceituais

O teste t de amostra única é logicamente idêntico ao z para uma média amostral. Ambos os testes calculam o desvio entre uma média amostral e algum valor esperado, que é uma média populacional ou um valor de interesse teórico para o pesquisador. Este desvio é o numerador do teste z ou teste t . O denominador de ambos os testes é um valor que representa o "erro amostral típico". A pequena diferença computacional entre o teste z e o teste t está no cálculo do erro típico de amostragem. A Tabela 7.1 destaca as semelhanças e diferenças entre esses dois testes de significâncias.

&nbsp; | z para uma média amostral | teste t de amostra única
-- | -- | --
Fórmula | $z = \frac{\bar{x}-\mu}{SEM_p}$ | $t =\frac{\bar{x}-\mu}{SEM_a}$
Cálculo do erro amostral (erro padrão da média) | $SEM_p = \frac{\sigma}{\sqrt{N}}$ | $SEM_a = \frac{DP}{\sqrt{N}}$

Ao usar o teste z, o desvio padrão da população é usado para calcular a quantidade esperada de erro amostral (ou seja, o denominador do teste, que é frequentemente chamado de erro padrão). Especificamente, o erro amostral é calculado usando a fórmula do erro padrão da média ($SEM_p = \sigma/\sqrt{N}$). Em algumas situações, entretanto, os pesquisadores podem não conhecer $\sigma$, o desvio padrão da população. Nessas situações, os pesquisadores calculam o desvio padrão dos dados amostrais e o utilizam para calcular uma estimativa do erro amostral esperado. Este procedimento é uma das diferenças entre o teste t de amostra única e o teste z. Especificamente, o erro de amostragem é calculado usando a fórmula para o erro padrão da média estimado ($SEM_a = DP/\sqrt{N}$) ao usar um teste t de amostra única.

Você deve se lembrar que as regiões críticas definem quais valores (valores z ou t) resultam na rejeição da hipótese nula. Outra diferença entre os testes z e t são os valores críticos usados para cada teste. Embora os valores críticos para os testes z sejam sempre os mesmos, os valores críticos para os testes t mudam com base no tamanho da amostra. Por exemplo, o valor crítico para um teste z unilateral com $\alpha = 0.05$ é sempre +1.65 ou -1.65. Dito de outra forma, os valores z de +1.65 ou -1.65 serão sempre os pontos que "delimitam" os 5% dos valores de z na região crítica dos 95% dos valores de z no corpo da distribuição. Por outro lado, os valores críticos para um teste t unilateral com $\alpha = 0.05$ mudarão com base em quantos elementos estão na amostra. A razão para isso é que a distribuição dos valores de t nem sempre tem formato normal. Quando o tamanho da amostra é pequeno, a curva dos valores t não tem o formato de uma curva normal. Na verdade, o formato da curva t é diferente para cada tamanho de amostra. A Figura 7.1 ilustra como cada tamanho de amostra cria uma curva t diferente, exibindo as diferentes curvas t para três tamanhos de amostra diferentes. A consequência importante das curvas t terem um formato diferente para cada tamanho de amostra é que os valores críticos que definem a região crítica são diferentes para cada tamanho de amostra (isto é, N). Portanto, estudos com tamanhos amostrais diferentes terão valores críticos diferentes, mesmo que ambos utilizem testes unicaudais com $\alpha = 0.05$. À medida que N aumenta, o valor crítico de t fica mais próximo do valor crítico de z.

````python
# HIDE CODE
# simular valores dentro da distribuição t
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
x=np.linspace(0-3*1, 0+3*1,1000)

T = stats.t(9)
tc = -T.ppf(0.05)
plt.plot(x, T.pdf(x), color='b', linestyle='dashed', 
         label=f'curva t com 10 amostras, valor crítico {round(tc, 4)}')

T = stats.t(30)
tc = -T.ppf(0.05)
plt.plot(x, T.pdf(x), color='black', label=f'curva t com 31 amostras, valor crítico {round(tc, 4)}')

T = stats.t(120)
tc = -T.ppf(0.05)
plt.plot(x, T.pdf(x), color='green', linestyle='dotted', 
         label=f'curva t com 121 amostras, valor crítico {round(tc, 4)}')

Z = stats.norm(0,1)
zc = -Z.ppf(0.05)
plt.plot(x, Z.pdf(x), color='red', linestyle='dashdot', 
        label=f'curva z normal padrão, valor crítico {round(zc, 4)}')

plt.legend(loc='upper center', bbox_to_anchor=(0, -0.05))
plt.title('A forma da distribuição t para diferentes tamanhos amostrais.')
#plt.savefig('comp_curvas_t.png', bbox_inches='tight', pad_inches=0.02, dpi=150)
````

## Exemplo do Teste t unicaudal de amostra única

Os estudantes universitários brasileiros estão contraindo dívidas de empréstimos estudantis através do FIES (Fundo de Financiamento ao Estudante do Ensino Superior). Em 2023, o Portal da Câmara dos Deputados estimou que o valor total da dívida federal de empréstimos estudantis no Brasil ultrapassaram 6.6 bilhões de reais. Embora existam muitas razões para o aumento da dívida de empréstimos estudantis, a pesquisa sobre uma heurística comum de tomada de decisão conhecida como viés de otimismo (Weinstein, 1980) sugere que um fator que contribui para o aumento da dívida de empréstimos estudantis pode ser o fato de os estudantes serem excessivamente otimistas sobre quais serão os salários após a formatura. Um conselheiro de ajuda financeira quer ver se os alunos estão excessivamente otimistas quanto aos seus salários futuros. Ele sabe que a distribuição dos salários iniciais dos formandos em farmácia que se formam em sua escola é normalmente distribuída com uma média de 36,000 Reais ou $\mu = 3000$ por mês. Para determinar se os estudantes de farmácia estão excessivamente otimistas em relação aos seus possíveis salários, a conselheira obtém uma amostra aleatória de 15 formandos em farmácia da sua escola e pergunta a cada um deles individualmente qual será o seu salário mensal no primeiro emprego após a formatura. Os salários iniciais mensais esperados dos 15 alunos eram 3500, 4200, 2500, 3200, 2700, 2000, 2500, 4100, 5000, 4200, 2900, 3900, 4000, 5000 e 5500, que da origem a uma média amostral de $\bar{x} = 3680$. Pesquisas anteriores sobre o viés do otimismo (por exemplo, Weinstein, 1980) sugerem que os alunos serão excessivamente otimistas nas suas estimativas e, portanto, o conselheiro opta por fazer um teste unilateral. Ela também escolhe um alfa de 0,05. O salário inicial médio estimado dos alunos é significativamente superior ao salário inicial real dos cursos de farmácia?

### _Etapa 1: examinar as pressuposições estatísticas_

O primeiro pressuposto de independência dos dados é cumprido porque o conselheiro recolheu os dados cuidadosamente, não permitindo que nenhum participante influenciasse as respostas dos outros participantes. A segunda suposição de medição apropriada da VI e da VD é atendida porque a VD, salário mensal esperado, é medido em como uma variável quantitativa, quanto mais (ou menos) cada participante espera ganhar, e a VI, expectativas dos estudantes de farmácia versus salários reais, identifica as expectativas salariais de um grupo de estudantes versus os salários reais de uma população. A terceira suposição de que a distribuição das médias amostrais terá uma forma normal também é satisfeita porque a população de salários esperados dos formandos em farmácia é normalmente distribuida. 
Ao realizar um teste t para amostra única, a última suposição de homogeneidade de variância é mais difícil de avaliar. A suposição é que o desvio padrão da amostra (DP) é semelhante ao desvio padrão da população ($\sigma$), mas quando você está fazendo um teste t de amostra única, você não conhece o desvio padrão da população ($\sigma$). Assim, os investigadores normalmente avaliam esta suposição utilizando o seu conhecimento das variáveis que estão estudando. O "tratamento" aplicado à amostra poderá alterar a variabilidade das pontuações? Neste estudo, o conselheiro está apenas perguntando aos alunos quanto dinheiro eles esperam fazer. Esta questão, se colocada com cuidado, é pouco provável que torne as respostas dos alunos mais ou menos variáveis, pelo que é seguro assumir que o desvio padrão da amostra será semelhante ao desvio padrão da população. No entanto, se o conselheiro conduzisse mal o estudo, talvez fazendo uma pergunta indutora como "Muitos estudantes sobrestimam quais serão os seus salários iniciais. Quanto você espera ganhar?" Fazer a pergunta desta forma poderia alterar as respostas dos alunos, fazendo com que cada aluno diminuísse a sua estimativa, o que poderia fazer com que as estimativas salariais da amostra fossem menos variáveis do que as estimativas salariais da população, o que violaria a suposição de homogeneidade de variância. Novamente, este exemplo ilustra que uma metodologia de pesquisa inadequada pode realmente atrapalhar suas estatísticas. Neste estudo, o conselheiro formulou a pergunta cuidadosamente para que a suposição de homogeneidade de variância fosse provavelmente atendida.

### _Etapa 3: Use o tamanho da amostra para calcular os graus de liberdade e defina as regiões críticas_

Na discussão anterior sobre regiões críticas, você aprendeu que à medida que o tamanho da amostra aumenta, o valor crítico se aproxima de zero. Para localizar a região crítica exata, você precisará do tamanho da amostra do estudo, bem como de uma tabela de valores t críticos (Apêndice B). Depois de saber o tamanho da amostra de um estudo, você precisa calcular seus graus de liberdade (gl). Para o teste t de amostra única, a fórmula gl é gl = N - 1. Portanto, neste caso,

$gl = 15 - 1 = 14$

Em seguida, você usa a tabela de valores t críticos no Apêndice B para encontrar o valor t crítico específico quando gl = 14. Nesse caso, você deve encontrar o gl de 14 na coluna mais à esquerda da tabela unicaudal e, em seguida, encontrar o valor crítico nessa linha sob o título 0.05. A Figura 7.2 ilustra como usar o Apêndice B.

````python
# HIDE CODE
import pandas as pd
import numpy as np
from scipy import stats

gls = [13, 14, 15]
a05 = [round(-stats.t(x).ppf(0.05), 4) for x in gls]
a01 = [round(-stats.t(x).ppf(0.01), 4) for x in gls]

df = pd.DataFrame(zip(gls, a05, a01), columns=['gl', r'$\alpha = .05$', r'$\alpha = .01$'])

color = (df.iloc[:,1] == df.loc[1,r'$\alpha = .05$']).map({True: 'background-color: yellow', False: ''})

df.style.apply(lambda s: color)
````

Neste caso, o valor t crítico é 1.7613. A hipótese de pesquisa previu que a média da nossa amostra seria maior do que a média da nossa população (ou seja, uma diferença positiva para $\bar{x} - \mu$). Assim, a região crítica está no lado positivo da distribuição, e a região crítica começa em +1.7613. Portanto, a hipótese nula deve ser rejeitada se o t obtido que calculamos na próxima etapa for igual ou superior a +1.7613, expresso matematicamente como t obtido $\geq +1.7613$.

### _Etapa 4: calcular a estatística do teste (teste t de amostra única)_
 
_4a. Calcule o desvio entre a média amostral e a média da população_

Comece calculando o desvio entre a média amostral e a média populacional. O salário médio mensal estimado dos 15 cursos de farmácia foi $\bar{x} = 3680$, e o salário mensal real foi $\mu = 3000$. Portanto, a diferença média é 680.

$\bar{x} - \mu =  3680-3000 = 680.$

_4b. Calcule o erro de amostragem esperado_

Para interpretar o desvio observado de 680, você precisa determinar quanta diferença você esperaria entre $\bar{x}$ e $\mu$ devido ao erro de amostragem. O erro de amostragem é estimado calculando a estimativa erro padrão da média estimado ($SEM_a = DP/\sqrt{N}$). Para calcular o erro de amostragem, você precisa conhecer o desvio padrão da amostra (DP) e tamanho da amostra (N). Neste problema, o desvio padrão da amostra não é fornecido a você. Portanto, você deve calculá-lo a partir dos dados de amostra. Como você deve se lembrar do Capítulo 3, calcular a variância (isto é, a média dos quadrados dos desvios) é o primeiro passo para calcular o DP (Desvio Padrão). A fórmula computacional para é $Var(X) = 1/(N-1)\sum_{i=1}^n(X_i - \bar{X})^2$. Obtivemos 15 valores, então N = 15. A _Var(x)_ é calculado conforme mostrado abaixo.
 
$Var(X) = \frac{1}{N-1}\sum_{i=1}^n(X_i - \bar{X})^2 = 1078857.143.$

O desvio padrão é calculado dividindo com o estimador não viesado da variância (com N - 1 no denominador) e extraindo a raiz quadrada do quociente.

E, finalmente, o erro padrão da média é calculado dividindo o desvio padrão pela raiz quadrada do tamanho da amostra.

$SEM_a = \frac{DP}{\sqrt{N}} = \frac{1038.680}{\sqrt{15}} = 268.186$
 
Neste caso, a quantidade típica de erro amostral foi calculada como 268.186. Portanto, as médias amostrais normalmente estarão 268.186 distantes da média populacional devido ao erro amostral.

_4c. Calcule a estatística do teste (teste t de amostra única)_

Dividindo a diferença observada entre a média amostral e a média populacional pela estimativa do erro amostral médio obtém-se o valor t calculado.

$t = \frac{(\bar{x} - \mu)}{SEM_a} = \frac{(3680-3000)}{268.186} = 2.54.$

Então, determinar se você deve rejeitar o nulo ou não é tão fácil quanto determinar se o valor t obtido (+2.54) está na direção prevista e se está mais distante de zero do que o valor crítico de +1.7613. Neste caso, o valor t obtido é positivo, conforme previsto, e é mais extremo que o valor crítico, portanto você tem evidências suficientes para rejeitar a hipótese nula.

Outra forma de descrever estes resultados é dizer: "O valor t obtido está na região crítica; portanto, há evidências suficientes para rejeitar a hipótese nula". Nesse caso, seu $\alpha = 0.05$ e seu t obtido estavam na região crítica. Isso significa que há menos de 5% de chance de você obter um valor t de +2,54 ou superior se a hipótese nula fosse verdadeira. Portanto, a análise estatística sugere que a hipótese nula é provavelmente falsa e que a hipótese de investigação é provavelmente verdadeira. Em outras palavras, o estudo sugere que os formandos em farmácia tendem a superestimar seus salários iniciais após a formatura.

### _Etapa 5: calcule o tamanho do efeito e descreva-o_

Tal como acontece com o z para uma média amostral, depois de concluir o teste de significância na Etapa 4, você deve calcular um tamanho de efeito para poder descrever a diferença entre as médias como leve, moderada ou grande. A estimativa do tamanho do efeito para o teste t de amostra única é calculada da mesma forma que foi para o z para uma média amostral, exceto que o desvio padrão da amostra (ou seja, DP) é o denominador, em vez do desvio padrão da população (ou seja, $\sigma$).

d = Desvio observado entre as médias/Desvio padrão $= \frac{(\bar{x} - \mu)}{DP} =  \frac{(3680 - 3000)}{1038.68} = .65.$

Um tamanho de efeito de 0.65 significa que os formandos em farmácia estimaram seu salário médio em 0.65 de um desvio padrão superior ao salário médio mensal real dos formandos em farmácia. As mesmas diretrizes de tamanho de efeito são usadas para z e t – ou seja, ds próximos a 0.2, 0.5 e 0.8 são efeitos pequenos, médios e grandes, respectivamente. Neste caso, o tamanho do efeito de 0.65 é de médio a grande, o que significa que os formandos em psicologia sobrestimam bastante os seus salários iniciais. No entanto, com um tamanho de amostra tão pequeno, é possível que este efeito seja resultado de um erro de amostragem.

### _Etapa 6: Interpretando os Resultados do Teste de Hipóteses_

Tal como antes com o z para a média amostral, o passo final é resumir os resultados, declarando explicitamente se os formandos em farmácia tendem a sobrestimar os seus salários iniciais ou não. Responder a esta pergunta foi a razão pela qual você fez todas essas estatísticas; é um desperdício de todo esse esforço se você não contar explicitamente às pessoas o que encontrou. Seu resumo também deve incluir a média da amostra e da população, os gl, o valor de t obtido, se o valor p foi menor ou maior que o valor alfa e o tamanho do efeito calculado. Um exemplo de descrição resumida é mostrado a seguir:

_O salário médio mensal esperado para formandos em farmácia ($\bar{x} = 680, DP = 1038.68$) foi significativamente maior do que o salário inicial real de 3.000 Reais, t(14) = 2.54, p < 0.05, d = 0.65. No capítulo anterior, você aprendeu a relatar o valor p exato em sua declaração resumida de a z para uma média amostral (por exemplo, p = 0.02). Se você souber o valor p exato de um resultado estatístico, deverá sempre relatá- lo. Ao usar um z para uma média amostral no capítulo anterior, você poderia facilmente determinar o valor exato de p para qualquer valor de z consultando-o na tabela normal padrão (ou seja, Apêndice A). No entanto, é mais difícil determinar o valor p exato de uma distribuição t porque, embora haja apenas uma probabilidade para cada valor de z, a probabilidade de uma valor t depende do tamanho da amostra do estudo. Portanto, não é prático para qualquer livro fornecer todas as probabilidades para cada valor de t e cada tamanho de amostra. Portanto, é comum simplesmente informar se o valor p é maior ou menor que o valor alfa escolhido (por exemplo, 0.05 ou 0.01) ao realizar um teste t manualmente. Se o valor t obtido estiver na região crítica, o valor p para esse valor de t deve ser menor que o valor alfa (0.05 ou 0.01). Portanto, você escreveria p < 0.05 sempre que rejeitasse o nulo. Por outro lado, se o valor de t obtido for menor que o valor crítico, seu valor p deve ser maior que o valor alfa e você escreveria p > 0.05. Para ser claro, a única razão pela qual você escreveria p < 0.05 ou p > 0.05 seria se você não soubesse o valor exato de p. Quando você calcula uma estatística manualmente, geralmente não saberá o valor exato de p dos seus resultados. No entanto, quando você usa um computador para realizar suas estatísticas, ele geralmente fornece o valor exato de p e, portanto, você deve relatá-lo.

## Exemplo de teste t de amostra únical de duas caudas (bicaudal)

No exemplo anterior, usamos um teste t unilateral (também conhecido como teste direcional). Também é possível usar um teste t bicaudal (também conhecido como teste não direcional). Por exemplo, suponha que, depois de saber que os formandos em farmácia eram excessivamente otimistas em relação aos seus próprios salários iniciais, você deseja saber se os alunos são igualmente otimistas ao estimar os salários iniciais dos outros. Para testar isso, você recruta 15 estudantes de farmácia diferentes e pede-lhes que estimem o salário médio mensal que um típico estudante de farmácia pode esperar em seu primeiro emprego após a formatura. Você não tem certeza se os alunos superestimarão ou subestimarão os salários iniciais dos outros e, portanto, você deve fazer um teste bicaudal em vez de um teste unilateral. Em testes de hipóteses bicaudais, você não tem uma previsão direcional específica em relação ao tratamento. Portanto, a hipótese de pesquisa bicaudal afirma que o salário médio mensal estimado será superior a 3000 Reais ou inferior a 3000. Assim, as hipóteses de pesquisa bicaudais possuem duas regiões críticas, uma na cauda positiva e outra na cauda negativa. Se o valor t obtido estivesse na região crítica positiva, você concluiria que os alunos superestimaram o salário inicial médio do típico do curso de farmácia. Por outro lado, se o valor t obtido estivesse na região crítica negativa, você concluiria que os alunos subestimaram o salário inicial médio. Finalmente, se o valor t obtido fosse "próximo de zero" (ou seja, entre os valores críticos positivos e negativos), você não conseguiria rejeitar a hipótese nula.
A escolha entre um teste unicaudal ou bicaudal pode ter um impacto dramático na(s) região(ões) crítica(s) e, portanto, sua decisão sobre usar um teste unicaudal ou bicaudal pode influenciar se você rejeita ou não a hipótese nula do teste. Por exemplo, se sua amostra tiver 15 pessoas e você realizar um teste unilateral com $\alpha = 0.05$, seu valor crítico seria +1.7613 e você rejeitaria o nulo se o valor t obtido fosse +1.8. No entanto, se você analisasse os mesmos dados, mas optasse por realizar um teste bicaudal, seus valores t críticos seriam +2.1448 e -2.1448. Se usar esse teste bicaudal, você não rejeitaria a hipótese nula se o valor de t obtido fosse +1.8 porque +1.8 não está em nenhuma das duas regiões críticas. A Figura 7.2 demonstra isso visualmente. A razão para as diferentes conclusões é que o valor crítico do teste unicaudal está mais próximo de zero do que os valores críticos bicaudais. Para ambos os tipos de testes, 5% da distribuição está em uma região crítica, mas com um teste bicaudal, 2.5% da distribuição está em cada cauda, enquanto com um teste unicaudal, 5% está em uma cauda única. Isso significa que o valor crítico estará mais próximo de zero para um teste unilateral do que para um teste bicaudal. Portanto, se a sua hipótese de pesquisa for direcional, é vantajoso usar um teste unilateral em vez de um teste bicaudal. Novamente, esse ponto é apresentado visualmente na Figura 7.3.

### _Etapa 3: Use o tamanho da amostra para calcular os graus de liberdade e definir as regiões críticas_

A fórmula para gl é a mesma para testes unicaudais e bicaudais. Como foi o caso no exemplo unilateral, o gl aqui é

$gl = N - 1= 15 - 1 = 14.$

Agora você usará a tabela de valores críticos de t no Apêndice B para determinar o valor crítico para um teste bicaudal quando gl = 14. Nesse caso, você encontra o gl na coluna mais à esquerda e então encontra o valor naquela linha sob o título bicaudal 0.05. O valor crítico é 2.1448. Portanto, as duas linhas de corte que definem as regiões críticas positivas e negativas estão localizadas em +2.1448 e -2.1448, respectivamente.

````python
# HIDE CODE
import pandas as pd
import numpy as np
from scipy import stats

gls = [13, 14, 15]
a05 = [round(-stats.t(x).ppf(0.05), 4) for x in gls]
a025 = [round(-stats.t(x).ppf(0.025), 4) for x in gls]

df = pd.DataFrame(zip(gls, a05, a025), columns=['gl', r'$\alpha = .05$', r'$\alpha = .025$'])

color = (df.iloc[:,1] == df.loc[1,r'$\alpha = .05$']).map({True: 'background-color: yellow', False: ''})

df.style.apply(lambda s: color)
````

### _Etapa 4: calcular a estatística do teste (teste t de amostra única)_

_4a. Calcule o desvio entre a média amostral e a média da população_

Esta etapa é idêntica a um teste unilateral:

$\bar{x} - \mu = 3160-3000 = 160.$

_4b. Calcule o erro médio de amostragem esperado_

Esta etapa é idêntica a um teste unilateral. Como no exemplo anterior, o desvio padrão da amostra (ou seja, DP) não é fornecido no problema, portanto você deve calculá-lo a partir dos dados da amostra. Se você não tiver certeza de como isso é feito, revise o exemplo anterior. Neste problema, o erro padrão da média é definido por

$SEM_a = \frac{DP}{\sqrt{N}}  = \frac{561.630}{\sqrt{15}} = 145.012$

Neste caso, a quantidade típica de erro amostral foi calculada em 145.012. A maneira correta de pensar sobre esse valor de erro amostral é que, embora algumas médias amostrais tenham mais erro amostral do que 145.012 e outras tenham menos, a quantidade típica de erro amostral para todas as amostras possíveis é 145.012.

_4c. Calcule as estatísticas do teste (teste t de amostra única)_

O cálculo da estatística de teste também é idêntico para testes unicaudais e bicaudais.

$t = \frac{(\bar{x} - \mu)}{SEM_a} = \frac{(3160 - 3000)}{145.012} = 1.10.$

O valor t obtido associado à média amostral de 3160 (ou seja, 1.10) não estava em nenhuma das regiões críticas. Isso significa que há mais de 5% de chance de que a diferença média no numerador se deva a um erro amostral. Portanto, você não deve rejeitar a hipótese nula. Os resultados sugerem que os formandos em farmácia não superestimam os salários iniciais dos demais. A Figura 7.5 pode ajudá-lo a visualizar o resultado desse teste de hipótese bicaudal. Como o valor t obtido não estava em nenhuma das regiões críticas, não há evidências suficientes para rejeitar a hipótese nula.

````python
# HIDE CODE
# simular valores dentro da distribuição normal N(0,1)
import matplotlib.pyplot as plt

x=np.linspace(0-3*1, 0+3*1,1000)

T = stats.t(14)
plt.plot(x, T.pdf(x), color='b')

tc = 2.145
tobs = 1.1

px=np.arange(tc,0+3*1,0.1)
plt.fill_between(px, T.pdf(px),color='r')

px=np.arange(-tc,0-3*1,-0.1)
plt.fill_between(px, T.pdf(px),color='r')

plt.text(2.13, 0.35, r'$t_c=2.145$')
plt.vlines(2.145, 0, 0.34)
plt.text(1, 0.35, r'$t_{obs}=1.1$')
plt.vlines(1.1, 0, 0.34)

plt.text(-2.13, 0.35, r'$t_c=-2.145$')
plt.vlines(-2.145, 0, 0.34)

plt.title('Distribuição t com região crítica bicaudal e t observado.')
#plt.savefig('regiao_critica_bicaudal_t_tobs.png', bbox_inches='tight', pad_inches=0.02, dpi=150)
````

## Outros Níveis de $\alpha$

Em ambos os exemplos anteriores, optamos por usar  $\alpha = 0.05$ porque em muitas áreas de pesquisa é o nível alfa mais comumente usado. No entanto, em alguns casos, você pode escolher o nível alfa mais rigoroso de 0.01. Um resultado do uso de $\alpha = 0.01$ é que a região crítica se torna menor. Por exemplo, se você usasse $\alpha = 0.01$ com um teste bicaudal com gl = 14, suas regiões críticas seriam $t \geq 2.9768$ e $t \leq -2.9768$. Se você usasse $\alpha = 0.01$ com um teste unilateral com gl = 14, suas regiões críticas seriam $t \geq 2.6245$.
Quando a hipótese nula é verdadeira, a probabilidade de um erro Tipo I é igual ao valor alfa que você escolher. Portanto, se você escolher o $\alpha = 0.01$ mais rigoroso, estará dificultando a rejeição da hipótese nula. Isso significa que é menos provável que você cometa um erro Tipo I, mas o nível alfa menor também afeta as taxas de erro Tipo II e o poder estatístico. O alfa menor de 0.01 significa que você tem menos poder estatístico e é mais provável que cometa um erro Tipo II do que se usar um valor alfa de 0.05. Assim, você precisa equilibrar seu desejo de minimizar os erros do Tipo I com seu desejo de ter poder estatístico (ou seja, sua capacidade de rejeitar a hipótese nula).
Neste livro, normalmente lhe diremos qual nível alfa usar, porque você ainda não tem informações suficientes sobre as questões de pesquisa para tomar uma decisão informada sobre qual nível alfa é mais apropriado. Os pesquisadores tomam essa decisão com base em vários fatores diferentes. Por exemplo, um pesquisador que está testando um tratamento para um distúrbio grave que atualmente não tem tratamento disponível ficará mais preocupado em cometer um erro do Tipo II (perder um tratamento realmente funciona) e usar um nível alfa mais alto (por exemplo, 0.05). ou mesmo 0.10). Os pesquisadores também podem escolher o nível alfa  com base no tamanho da amostra. Por exemplo, um investigador que estuda as diferenças de género nas pontuações dos testes de matemática pode ter acesso às pontuações de milhões de crianças. Neste caso, o pesquisador provavelmente usaria um nível alfa muito mais baixo (por exemplo, 0.01 ou mesmo 0.001) porque tem tanto poder estatístico que não deseja relatar efeitos estatisticamente significativos, mas praticamente sem sentido (ou seja, têm tamanhos de efeito muito pequenos).
